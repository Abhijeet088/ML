```python
from sympy import symbols , diff
print("Sympy imported successfully")
```

```python
w, x, b = symbols('w x b ')
print("Defined symbols")
print("w =",w )
print("x=" , x )
print("b =", b )
```

```python
z = w*x + b
J = z**2
print("z = ",z)
print("J = ", J)
```

```python
dj_dw = diff(J, w)
print("Derivative of J with respect to w: ")
print("dj/dw =", dj_dw)
```

```python
w_val = 3
x_val = 2
b_val = 1
z_val = w_val*x_val + b_val
J_val = z_val** 2
print("forward Pass result: ")
print("z=", z_val )
print("j = ", J_val )
```

```python
dj_dz = 2*z_val
print("dj/dz = ", dj_dz)
dz_dw = x_val
print("dz_dw =", dz_dw)
dj_dw_manual = dj_dz *dz_dw
print("Manual Backprop Result:")
print("dj/dw =" , dj_dw_manual )
```
